```python
import nltk, re
import urllib.request
```

```python
def get_players_from_club(inputCountry, inputLeague, inputClub) :
    #Gets club info
    response = urllib.request.urlopen("https://www.bold.dk/fodbold/{}/{}/{}/".format(inputCountry, inputLeague, inputClub))
    data = response.read()
    clubText = data.decode('Latin-1')
    
    #Finds the players in the team
    matches = re.findall(r"<h2>Spillertrup.*</h2>\n<table class=\"personlist\".*>((.|\n)*?)</tr></table>", clubText)
    teamPlayersText, b = matches[0]
    
    #Finds details(clubNationality, league, club, name, position) for each player
    players = re.findall(r"href=\"/.*?/(.*?)/(.*?)/(.*?)/(.*?)/\"(.|\n)*?class=\"small position\">(.*?)<", teamPlayersText)
    return players
```

```python
#For each player in a club, gather more details (Nationality, age, height, weight)
#And then write the infor to a file
def write_player_info_for_club(inputCountry, inputLeague, inputClub) :
    players = get_players_from_club(inputCountry, inputLeague, inputClub)
    
    f = open("playerDataNew.csv","a+")
    for clubNationality, league, club, name, b, position in players :
        
        playerURL = "https://www.bold.dk/fodbold/{}/{}/{}/{}".format(clubNationality, league, club, name)
        #print("{} - {}".format(club,name))
        response = urllib.request.urlopen(playerURL)
        data = response.read()
        text = data.decode('Latin-1')

        line = ""
        m = re.findall(r"itemprop=\"image flag\".*?>.*?(\w*?)<(.|\n)*?birthDate.*\n.*?\((\w*).*?/td>(.|\n)*?jde(.|\n)*?td>(\w*)(.|\n)*?gt(.|\n)*?td>(\w*)", text)
        if not m :
            m = re.findall(r"itemprop=\"image flag\".*?>.*?(\w*?)<(.|\n)*?birthDate.*\n.*?\((\w*).*?/td>", text)
            if not m :
                line = "{},{},{},{},{},{},{},{},{}\n".format(name, position, nationality, "-", "-", "-", inputClub, inputCountry, inputLeague)

        if line == "" :
            playerData = m[0]
        try :
            nationality, c, age, d, e, height, g, h, weight = playerData
            line = "{},{},{},{},{},{},{},{},{}\n".format(name, position, nationality, age, height, weight, inputClub, inputCountry, inputLeague)
        except :
            nationality, c, age = playerData
            line = "{},{},{},{},{},{},{},{},{}\n".format(name, position, nationality, age, "-", "-", inputClub, inputCountry, inputLeague)
        f.write(line)

    f.close()
```
